```python
import re
import regex


begin_brackets = [
    "{|",
    "{{",
    "[["
]

end_brackets = [
    "|}",
    "}}",
    "]]",
]

bracket_begin_regex = "({})".format("|".join([re.escape(s) for s in begin_brackets]))
bracket_end_regex = "({})".format("|".join([re.escape(s) for s in end_brackets]))
bracket_all_regex = "({})".format("|".join([re.escape(s) for s in begin_brackets + end_brackets]))

#parenthesis_regex = r"(?<rec>{begin}(?:((?!({begin}|{end})).)*|(?&rec))*{end})".format(begin=bracket_begin_regex, end=bracket_end_regex)

parenthesis_regex = r"(?<rec>{begin}(?:((?!{all}).)*|(?&rec))*{end})".format(begin=bracket_begin_regex, end=bracket_end_regex, all=bracket_all_regex)

parenthesis_exp = regex.compile(parenthesis_regex, flags=regex.DOTALL)
```
